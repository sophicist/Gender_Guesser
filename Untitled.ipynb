```python
import pandas as pd
from functions import namer,namer1
import numpy as np
```

```python
F1 = "john"
F2 = "kamau"
```

```python
def chooser(F1,F2):
    X =namer(F1)
    Y =namer1(F2)
    if X[1]>Y[1]:
        return(X[0])
    else:
        return(Y[0])
```

```python
chooser(F1,F2)
```

```python
df = pd.read_csv('mergedfinalstats.csv')
df["Merchant's Name"].head()
```

```python
x = df[df.columns[df.columns.str.contains("Gender")]]
dt = pd.DataFrame(df["Merchant's Name"])
dt['gender'] = pd.Series(x.columns[np.where(x!=0)[1]])
dt['gender'] = list(map(lambda x:str(x).replace('Gender_','').replace(',',""),dt['gender']))
dt.head()
```

```python
def clean(name):
    name = str(name).lower().replace(',',"").split(' ')
    try:
        choose = chooser(name[0],name[1])
        return choose.capitalize()
    except:
        return 'error'
dt['Guess'] = list(map(clean,dt["Merchant's Name"]))
```

```python
dt.head()
```

```python
print(dt.gender.value_counts())
print(dt.Guess.value_counts())
```

```python
from sklearn.metrics import confusion_matrix as cm,accuracy_score as ac
dt[dt.Guess =='error'] =np.nan
dt = dt.dropna()
print(cm(dt['gender'],dt['Guess']))
print(ac(dt['gender'],dt['Guess']))
```
